```python
ot=[]
list_of_teams1=[]
for df in list_of_teams:

  column_num = 0
  column_np = df.iloc[:, column_num].values
  ot.append(column_np)
  df = df.drop(df.columns[column_num], axis=1)
  list_of_teams1.append(df)
```

```python
import pandas as pd
import numpy as np





for X in list_of_teams1:
  X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)


for X in list_of_teams1:
  X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)


theta=[]
for X in list_of_teams1:
  thetak = np.zeros(X.shape[1])
  theta.append(thetak)
print(thetak)

learning_rate = 0.01
num_iterations = 1000
BIAS=[]
for i in range(len(list_of_teams1)):
  k=list_of_teams1[i]
  M=ot[i]
  Bias=0

  thetak=theta[i]

  for iteration in range(num_iterations):


    for j in range(len(M)):
      Y=M[j]
      X=k.iloc[j]

      errors=0


      predictions = np.dot(X, thetak).sum()

      if iteration==0:
          errors= predictions - Y
      else:
          errors= errors +predictions - Y



    gradients = (2 / len(X)) * np.dot(X.T, errors).sum()
    grad_b = (-2 / len(X)) * np.sum(errors)

    thetak -= learning_rate * gradients
    Bias += learning_rate*grad_b


    final_predictions = np.dot(X, thetak).sum()+Bias

            # Calculate MSE
    mse = np.mean(errors)
    print("Mean Squared Error (MSE):",iteration,final_predictions,mse)
  BIAS.append(Bias)
```

```python
def gradient(x,y,theta,b):
    
    alpha = 0.01
    iteration = 1000
    #gradient descend algorithm
    J_history = {}
    for iter in range (iteration):
        
        error = (x @ theta) + b -y
        for i in range (np.shape(x)[1]):
          theta[i] = theta[i] - ((alpha) * np.mean(error*x[:,i]))
        b = b - ((alpha) * np.mean(error))
        if iter % 100 == 0:
          J_history[iter] = ((0.5) * (np.mean(((x @ theta + b)-y)**2)))   #compute J value for each iteration
    return theta, J_history, b
```

#Making Predictions <br>
Now that we have trained our model using given data, as well as validated it, we will proceed to make our predictions

```python
test = pd.read_csv('/content/Test_dataset.csv')
test.drop(['stadium', 'place'], axis=1, inplace=True)
test
```

```python
my_dict = {}
my_dict['Chennai Super Kings'] = 0
my_dict['Delhi Capitals'] = 1
my_dict['Kolkata Knight Riders'] = 2
my_dict['Mumbai Indians'] = 3
my_dict['Punjab Kings'] = 4
my_dict['Rajasthan Royals'] = 5
my_dict['Royal Challengers Bangalore'] = 6
my_dict['Sunrisers Hyderabad'] = 7

print (my_dict['Chennai Super Kings'])
```

```python
result = []

for i in range (45):
  a = test.iloc[i][1]
  b = test.iloc[i][2]
  key_a = (my_dict[b])
  key_b = (my_dict[a])

  #calculating 'a' prediction with b opponent
  params = []
  params.append(1)
  if test.iloc[i][3] == a:
    params.append(1)
  else:
    params.append(0)

  if (test.iloc[i][3] == a and test.iloc[i][4] == 'Field') or (test.iloc[i][3] == b and test.iloc[i][4] == 'Bat'):
    params.append(0)
  else:
    params.append(1)
  theta = df.loc[df['Team'] == a].iloc[0][1]
  add = df.loc[df['Team'] == a].iloc[0][2]
  hot = df.loc[df['Team'] == a].iloc[0][3]
  hot = hot[key_a]
  params.extend(hot)
  target_a = (params @ theta) + add

  #calculating 'b' prediction with a as opponent
  params = []
  params.append(1)
  if test.iloc[i][3] == b:
    params.append(1)
  else:
    params.append(0)

  if (test.iloc[i][3] == b and test.iloc[i][4] == 'Field') or (test.iloc[i][3] == a and test.iloc[i][4] == 'Bat'):
    params.append(0)
  else:
    params.append(1)
  theta = df.loc[df['Team'] == b].iloc[0][1]
  add = df.loc[df['Team'] == b].iloc[0][2]
  hot = df.loc[df['Team'] == b].iloc[0][3]
  hot = hot[key_b]
  params.extend(hot)
  target_b = (params @ theta) + add

  if target_a > target_b:
    result.append(1)
  elif target_a < target_b:
    result.append(2)
  else:
    result.append(0)

print(result)
```

```python
Final = pd.DataFrame()
Final['team1'] = test['team1']
Final['team2'] = test['team2']
Final['prediction'] = result
Final
```

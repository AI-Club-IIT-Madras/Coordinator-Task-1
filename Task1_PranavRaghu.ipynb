### Data Preprocessing and Training

```python
def Predict(team, raw_df, test_df):
    team_df = raw_df.loc[((raw_df['team1'] == team) | (raw_df['team2'] == team))]
    raw_df = raw_df.reset_index()

    team_pred = test_df.loc[((test_df['team1'] == team) | (test_df['team2'] == team))]
    test_df = test_df.reset_index()

    team_scores = team_df.apply(lambda row: row['team1_score'] if row['team1'] == team else row['team2_score'], axis=1)
    team_scores = team_scores.values.tolist()
    team_scores = np.array(team_scores).reshape(1,-1)

    # Make an opponents list for the team
    team_opponents = team_df.apply(lambda row: row['team2'] if row['team1'] == team else row['team1'], axis=1)
    team_opponents = team_opponents.values.tolist()

    test_opponents = team_pred.apply(lambda row: row['team2'] if row['team1'] == team else row['team1'], axis=1)
    test_opponents = test_opponents.values.tolist()

    team_df = team_df.assign(Opponent=team_opponents)
    team_df.drop(['team1','team2','team1_score','team2_score'], inplace=True, axis=1)
    team_df.reset_index(drop=True, inplace=True)

    team_pred = team_pred.assign(Opponent=test_opponents)
    team_pred.drop(['team1','team2'], inplace=True, axis=1)
    team_pred.reset_index(drop=True, inplace=True)

    # Make a list that reflects 1 for team batting first and 0 for team batting second
    team_batting = []
    for i in range(len(team_df)):
        if team_df['toss_winner'][i] == team:
            if team_df['toss_choice'][i] == 'Bat':
                team_batting.append(1)
            else:
                team_batting.append(0)
        else:
            if team_df['toss_choice'][i] == 'Field':
                team_batting.append(1)
            else:
                team_batting.append(0)

    test_batting = []
    for i in range(len(team_pred)):
        if team_pred['toss_winner'][i] == team:
            if team_pred['toss_choice'][i] == 'Bat':
                test_batting.append(1)
            else:
                test_batting.append(0)
        else:
            if team_pred['toss_choice'][i] == 'Field':
                test_batting.append(1)
            else:
                test_batting.append(0)

    team_df['Toss Outcome'] = team_batting
    team_df.drop(['toss_winner','toss_choice'], inplace=True, axis=1)
    team_df = team_df.reset_index(drop=True)
    
    team_pred['Toss Outcome'] = test_batting
    team_pred.drop(['toss_winner','toss_choice'], inplace=True, axis=1)
    team_pred = team_pred.reset_index(drop=True)

    # Number Encoding for the year column
    team_df['year'] = team_df['year'].apply(lambda x: (x-2000)/22)
    team_pred['year'] = team_pred['year'].apply(lambda x: (x-2000)/22)

    # One hot encoding
    temp = pd.get_dummies(team_df, columns=['Opponent', 'stadium'])
    temp.drop(['Toss Outcome','year'], inplace=True, axis=1)
    team_df = pd.concat([team_df, temp], axis=1)
    team_df.drop(['Opponent', 'stadium'], inplace=True, axis=1)
    team_df = team_df.reset_index(drop=True)

    temp = pd.get_dummies(team_pred, columns=['Opponent', 'stadium'])
    temp.drop(['Toss Outcome','year'], inplace=True, axis=1)
    team_pred = pd.concat([team_pred, temp], axis=1)
    team_pred.drop(['Opponent', 'stadium'], inplace=True, axis=1)
    team_pred = team_pred.reset_index(drop=True)

    # Initialising X, weights and biases
    X = team_df.to_numpy()
    n = X.shape[0]

    # Parameters
    in_dim = X.shape[1]
    out_dim = 1
    
    # Xavier Initialisation
    limit = np.sqrt(6 / float(in_dim + out_dim))
    W = np.random.uniform(low=-limit, high=limit, size=(in_dim, out_dim))
    b = 0
        
    # Hyperparameters
    learning_rate = 0.03
    num_iterations = 5000
    costs = []

    for i in range(num_iterations):
        # Forward Propagation
        y = np.dot(X, W) + b

        # MSE Loss
        loss = np.mean((y.T - team_scores)**2)

        # Computing gradients
        dw = (2 / n) * np.dot(X.T, (y.T - team_scores).T)
        db = (2 / n) * np.sum(y.T - team_scores)

        # Backpropagation
        W = W - learning_rate * dw
        b = b - learning_rate * db

        costs.append(loss)

    plt.plot(costs)
    

    # Assign team_pred to the intersection of team_pred and team_df and remove all columns except the ones that are needed for prediction
    team_pred = team_pred.loc[:,team_pred.columns.intersection(team_df.columns)]

    temp = []
    for i in range(len(team_df.columns)):
        if team_df.columns[i] in team_pred.columns:
            temp.append(W[i][0])
    W = temp
    W = np.array(W)

    # Predicting the scores
    temp = team_pred.to_numpy()
    predicted_score = np.dot(W.T, temp.T) + b
    team_predictions = []

    j = 0
    for i in range(len(test_df)):
        if test_df['team1'][i] == team:
            team_predictions.append([test_df['team2'][i], test_df['stadium'][i], predicted_score[j]])
            j+=1
        elif test_df['team2'][i] == team:
            team_predictions.append([test_df['team1'][i], test_df['stadium'][i], predicted_score[j]])
            j+=1

    return team_predictions
```

```python
df = raw_df.copy()
df.drop(columns =['margin','man_of_the_match','winner','place'], axis=1, inplace=True)
df = df.dropna()
df.reset_index(inplace = True)
# print(teams)
dataframes = dict()
for team in teams:
    team_rows = df[(df['team1'] == team) | (df['team2'] == team)]
    team_rows.reset_index(inplace=True)
    # if team == "Royal Challengers Bangalore":
    #     print(team_rows)
    scores = list()
    opponent_teams = list()
    # print(team_rows.team2_score.values[5])
    for i in range(len(team_rows)):
        # if i==1:continue
        score = (team_rows.team1_score.values[i] if team_rows.team1.values[i] == team else team_rows.team2_score.values[i])
        scores.append(score)
        # print(team, team_rows.toss_winner[i])
        team_rows["toss_winner"][i] = (1 if (team_rows["toss_winner"][i] == team) else 0)
        team_rows["toss_choice"][i] = (1 if (team_rows["toss_choice"][i] == "Bat") else 0)
        opponent_teams.append(team_rows["team1"][i] if team_rows["team2"][i] == team else team_rows["team2"][i])
    team_rows.drop(columns =['team1_score','team2_score','index'], axis=1, inplace=True)
    # print((opponent_teams))
    team_rows["team_score"] = scores
    team_rows["Opponent_Team"] = opponent_teams
    team_rows.drop(columns=["team1","team2", "level_0"], axis=1, inplace=True)
    team_rows.rename(columns={"toss_choice":"chose_to_bat"}, inplace=True)
    dataframes[team] = [team_rows]

    # print(team,scores)
```

```python
# Mean square loss function
def loss(outputs, targets):
    # print(outputs.shape)
    loss = np.sum(np.power(outputs-targets, 2))/outputs.shape[0]
    return loss
def gradient(inputs, targets, weights, outputs=None):
    if not outputs: outputs = output(inputs, weights)
    gradient = (np.sum((outputs-targets)*inputs, axis=0))/outputs.shape[0]
    gradient = gradient.reshape(gradient.shape+(1,))
    return gradient
print(df.shape[1])
```

```python
def train(inputs, targets, weights, step_size=0.1, no_of_epochs=3000):
    for i in range(no_of_epochs):
        outputs = output(inputs, weights)
        weights = weights - (gradient(inputs, targets, weights))*step_size
    return weights, loss(outputs, targets)
        
        
```

```python
def validate(raw_df, columns_to_drop=['margin','man_of_the_match','place','index'], winner_known = True, year=2021):
    df = raw_df.copy()
    df = df.dropna()
    df.reset_index(inplace=True)
    df = df[df["year "]==year]
    df.drop(columns =columns_to_drop, axis=1, inplace=True)
    right_guesses = 0
    predictions = []
    for i in range(len(df)):
        team1 = df.loc[i]["team1"]
        team2 = df.loc[i]["team2"]
        toss_winner1 = 1 if df.loc[i]["toss_winner"] == team1 else 0
        toss_winner2 = 1 - toss_winner1
        stadium = df.loc[i]["stadium"]
        if winner_known: winner = df.loc[i]["winner"]
        encoded_opponents1 = [1 if opp == team2 else 0 for opp in encoding[team1]["opponent_"]]
        encoded_opponents2 = [1 if opp == team1 else 0 for opp in encoding[team2]["opponent_"]]
        encoded_stadiums1 = [1 if s == stadium else 0 for s in encoding[team1]["stadium_"]]
        encoded_stadiums2 = [1 if s == stadium else 0 for s in encoding[team2]["stadium_"]]
        inputs_team1 = np.array([toss_winner1] + encoded_opponents1 + encoded_stadiums1)
        inputs_team2 = np.array([toss_winner2] + encoded_opponents2 + encoded_stadiums2)
        pred_team1_score = output(inputs_team1,dataframes[team1][2])
        pred_team2_score = output(inputs_team2,dataframes[team2][2])
        pred_winner = team1 if pred_team1_score > pred_team2_score else team2
        if pred_winner == team1: predictions.append(1)
        else: predictions.append(2)
        if winner_known:
            correct = winner == pred_winner
            if correct: right_guesses +=1 
    if winner_known:
        accuracy = right_guesses/len(df)
        return accuracy
    return predictions
validate(raw_df)
```
